### Model Building

```python
#Split Data
X_Data=californiaHousingDf[['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income','ocean_proximity_num']]
Y_Data=californiaHousingDf["median_house_value"]
```

The proportion of data split between the training set and the test set is a crucial decision when using the train-test split method for model evaluation in machine learning. The most common split ratios are 70-30, 80-20, or 75-25, but the choice depends on factors such as the size of your dataset, the complexity of your model, and the specific goals of your analysis.

```python
#Here I am allocating 33 percent as Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Data, Y_Data, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Model = LinearRegression()
```

```python
#Trained My Model
Model.fit(X_train,y_train)
```

```python
print("It represents the estimated value of the target variable when all features are zero", Model.intercept_)
print(Model.coef_)
```

```python
Coeff_Frame = pd.DataFrame(data=Model.coef_,index=X_train.columns,columns=['Coeff'] )
Coeff_Frame
```

#### Predictions

```python
Predictions = Model.predict(X_test)
```

```python
Predictions
```

```python
#To visualize if we had made a starigth line model
plt.scatter(y_test,Predictions)
plt.show()
```

Let's Make Histogram of Residuals to see more clearly our Model picture.  Residuals are difference betwen Actual Value and Prediction Values.

#### Regresion Evaluation Metrics

The loss function quantifies how well the model's predictions match the actual target values. It measures the "loss" or error between the predicted values and the true values, providing a measure of how far off the model's predictions are from the actual outcomes.<br>
If prediction accuracy is crucial, Then in Regression task MSE or RMSE may be more appropriate.Now lets calculate loss function or efficiency of our model. 

```python
from sklearn import metrics
def printModelAccuracy(y_test,p):
    #Mean Absolute Error
    print("MAE: ",metrics.mean_absolute_error(y_test,p))
    #Mean Square Error, More popular as it punishes larger error, as it squares.
    print("MSE: ",metrics.mean_squared_error(y_test,p))
    #Square Root Mean square Error
    print("RMSE: ",np.sqrt(metrics.mean_squared_error(y_test,p)))
    #Goodness of a Linear Fit from 0 to 1, but there is chance of higher value with more independent variables, in that case see Adjusted R-sQUARE
    print("R-Square: ",metrics.r2_score(y_test,p))
```

```python
printModelAccuracy(y_test,Predictions)
```
